```python
import os
import re

import pandas as pd
```

```python
path = '/Users/leportella/src/lele/leportellav3/_posts/'
```

```python
files = [f for f in os.listdir(path)]
```

```python
infos = []
for file in files:
    f = open(path + file)
    lines = f.readlines()
    
    year = file.split('-')[0]
    month = file.split('-')[1]

    categories = []
    translated = False
    for line in lines:
        if line.startswith('title'):
            title = re.match('title: "(.*)"', line).groups(0)[0]
        if line.startswith('categories'):
            category = True
        if line.startswith('tags'):
            category = False
        if line.startswith('  -') and category:
            cat = re.match("  - (.*)\\n", line).groups(0)[0]
            categories.append(cat.strip())
        if line.startswith("translation"):
            translated = True
    
    lang = categories[0]
    is_interview = 'interview' in categories
    is_talk = 'talks' in categories
    is_handbook = 'handbook' in categories
    is_cheatsheet = 'cheatsheet' in categories
    is_post = len(categories) == 1

    infos.append([year, month, title, lang, categories, translated, is_post, is_cheatsheet, is_interview, is_talk, is_handbook])
```

```python
columns = ['year', 'month', 'title', 'lang', 'categories', 'is_translated', 'is_post', 'is_cheatsheet', 'is_interview', 'is_talk', 'is_handbook']
```

```python
df = pd.DataFrame(infos, columns=columns)
```
